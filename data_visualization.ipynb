```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
data = pd.read_excel('TSIC hackathon dataset.xlsx')
data.head()
```

```python
data[data['Female population age 6 years and above who ever attended school (%)']>50].head()

```

```python
data[data['Female population age 6 years and above who ever attended school (%)']>60].head()
     
```

```python
plt.figure(figsize=(60,10))
plt.bar(data['Indicators (2019-20)'],data['Female population age 6 years and above who ever attended school (%)'])
plt.show()
```

```python
plt.figure(figsize=(100,60))
data.groupby(['Indicators (2019-20)'])[['Female population age 6 years and above who ever attended school (%)', 'Women age 20-24 years married before age 18 years (%)']].sum().plot(kind = "line")
```

```python
plt.figure(figsize=(100,60))
data.groupby(['Indicators (2019-20)'])[['Child Feeding Practices and Nutritional Status of Children', 'Children under 5 years who are underweight (weight-for-age)18 (%)']].sum().plot(kind = "line")
```

```python
plt.figure(figsize=(100,60))
data.groupby(['Indicators (2019-20)'])[['Women age 20-24 years married before age 18 years (%)', 'Women whose Body Mass Index (BMI) is below normal (BMI <18.5 kg/m2)21 (%)', 'Pregnant women age 15-49 years who are anaemic (<11.0 g/dl)22 (%)']].sum().plot(kind = "line")
```

```python
plt.figure(figsize=(100,60))
data.groupby(['Indicators (2019-20)'])[['Female population age 6 years and above who ever attended school (%)', 'Women age 15 years and above who use any kind of tobacco (%)']].sum().plot(kind = "bar")
```
